# Импортируем все, что нам нужно:

```python
from ludwig.api import LudwigModel
import pandas as pd
pd.__version__
```

# Data:

```python
data = pd.read_csv ('ks-projects-201801.csv', encoding = "ISO-8859-1", engine='python')
data. columns
```

```python
data1 = pd.DataFrame(data)
```

```python
data.head()
```

# Посмотрим, насколько зависит успешность мероприятия от инвестиций и его категории

# Строим модель:

```python
model_definition = {
    "input_features": [
        {'name': 'goal', "type": "numerical"},
        {"name": 'main_category', "type": "category"},
        {"name": 'backers', "type": "numerical"}
    ],
    "output_features":[ 
        {"name": 'state', "type":  'category'}
        ] 
}
model = LudwigModel(model_definition)
train_stats = model.train(data)
```

# Тренируем и смотрим на точность:

```python
train = model.train(data1)
predictions = model.predict(data1)
```
